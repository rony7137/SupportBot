```python
import json
#read the corpus
with open('GL Bot.json') as file:
  Corpus = json.load(file)

print(Corpus)
```

```python
Corpus['intents'][0]
```

```python
#create the list of tokens and tags

import nltk
nltk.download('punkt')

token_list = [] #to store the tokens
tag_list = [] #to store the tags
sentence = [] #to store the patterns
tag = [] #to store the corresponding tag

for intent in Corpus['intents']:
  for pattern in intent['patterns']:
    w = nltk.word_tokenize(pattern)
    token_list.extend(w)
    sentence.append(w)
    tag.append(intent['tag'])

  if intent['tag'] not in tag_list:
    tag_list.append(intent['tag'])
```

```python
# convert to lower case & stemming
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer('english')
token_list = [stemmer.stem(w.lower()) for w in token_list]
token_list = sorted(set(token_list))

tag_list = sorted(tag_list)
```

```python
print(len(token_list))
print(token_list)
```

```python
print(len(tag_list))
print(tag_list)
```

```python
#Create the bag of words for the input & one-hot encoding for the target variables

X = []
Y = []

out_temp = [0 for _ in range(len(tag_list))]

#loop to create the bag of words
for i, text in enumerate(sentence) :
  bag = []
  w = [stemmer.stem(w.lower()) for w in text]

  for tk in token_list:
    if tk in w:
      bag.append(1)
    else:
      bag.append(0)
#create the one-hot encoded output
  out = out_temp[:]
  out[tag_list.index(tag[i])] = 1

  X.append(bag)
  Y.append(out)
```

```python
print(X[0])
```

```python
print(Y[0])
```

```python
# NN model for prediction of the tag

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

model = Sequential()
model.add(Dense(64, input_dim = len(X[0]), activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(len(Y[0]), activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```

```python
model.fit(X, Y, epochs=200, batch_size=30)
```

```python
# define a function to convert the input text to bag of words

def bag_of_words(text):
  words = nltk.word_tokenize(text)
  
  bag = []
  w = [stemmer.stem(w.lower()) for w in words]

  for tk in token_list:
    if tk in w:
      bag.append(1)
    else:
      bag.append(0)
  return bag
```

```python
# Test prediction
results = model.predict([bag_of_words('Hello')])
results_index = np.argmax(results)
pred = tag_list[results_index]
pred
```

```python
# define a chat function

def chat():
  print("Hi my name is Jason")

  while True:
    ip = input("\n\nYou:")
    if ip.lower() == '?':
      print("BOT: Please rephrase your question and try again")
    if ip.lower() == 'quit':
      break

    results = model.predict([bag_of_words(ip)])
    results_index = np.argmax(results)
    tag_pred = tag_list[results_index]

    for intent in Corpus['intents']:
      if intent['tag'] == tag_pred:
        responses = intent['responses']

    print(np.random.choice(responses))
```

```python
chat()
```
